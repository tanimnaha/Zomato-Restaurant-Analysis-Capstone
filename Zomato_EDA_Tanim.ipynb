## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# --- 1. CLEANING METADATA ---
# Fix the 'Cost' column: remove commas and turn it into a number
df_meta['Cost'] = df_meta['Cost'].astype(str).str.replace(',', '')
df_meta['Cost'] = pd.to_numeric(df_meta['Cost'])

# Fill missing values in Metadata with 'Not Available'
df_meta['Collections'] = df_meta['Collections'].fillna('Not Available')
df_meta['Timings'] = df_meta['Timings'].fillna('Not Available')

# --- 2. CLEANING REVIEWS ---
# Drop duplicate rows in Reviews
df_reviews = df_reviews.drop_duplicates()

# Drop rows that have missing values in Reviews
df_reviews = df_reviews.dropna()

# --- 3. VERIFY THE CLEANUP ---
print("Metadata missing values now:\n", df_meta.isnull().sum())
print("\nReviews missing values now:\n", df_reviews.isnull().sum())
print("\nReviews rows remaining after cleanup:", df_reviews.shape[0])
````

### What all manipulations have you done and insights you found?

Manipulations Done:

Data Type Conversion: I removed the commas from the Cost column in the Metadata dataset and converted it into a numeric data type so it can be used for calculations and charts.

Handling Missing Values (Imputation): I filled the missing values in the Collections and Timings columns with the string "Not Available" to ensure no blank spaces remained.

Handling Duplicates and Nulls (Deletion): In the Reviews dataset, I dropped the duplicate rows. I also dropped the rows containing missing values (nulls) because they represented a tiny fraction of the massive 10,000-row dataset, meaning their removal will not negatively impact our analysis.

Insights Found:

The datasets are now completely clean with 0 missing values and 0 duplicates. The Reviews dataset retained 9,955 clean rows. We are now ready to begin creating visual charts for Exploratory Data Analysis (EDA).

````python
# Chart - 9 visualization code
import matplotlib.pyplot as plt
import seaborn as sns

# Extract the 'Followers' count from the text in the Metadata column
df_reviews['Followers'] = df_reviews['Metadata'].str.extract(r'(\d+)\s+Follower').astype(float).fillna(0)

# Find the top 10 reviewers with the most followers
top_critics = df_reviews.groupby('Reviewer')['Followers'].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_critics.values, y=top_critics.index, palette='flare')
plt.title('Top 10 Zomato Critics by Follower Count')
plt.xlabel('Number of Followers')
plt.ylabel('Reviewer Name')
plt.show()
````

##### 2. What is/are the insight(s) found from the chart?

The chart successfully identifies the top "critics" or influencers on the platform. A few specific reviewers have amassed massive followings (some well over thousands of followers). This indicates that Zomato has a micro-influencer community where everyday users look to specific established critics for dining recommendations.

However, the largest surge in activity occurs at night, peaking heavily between 20:00 and 23:00 (8:00 PM to 11:00 PM), right after dinner. Very few reviews are posted in the early morning hours.

````python
# Correlation Heatmap visualization code
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Let's quickly re-merge our tables so it includes the 'Hour' column we made in Chart 11!
merged_df = pd.merge(df_reviews, df_meta, left_on='Restaurant', right_on='Name', how='inner')

# Grab only the numerical columns for a correlation heatmap (.copy() prevents warnings)
numeric_cols = merged_df[['Cost', 'Rating', 'Pictures', 'Hour']].copy()

# For the review stats, we extract them fresh
numeric_cols['Total_Reviews'] = merged_df['Metadata'].str.extract(r'(\d+)\s+Review').astype(float).fillna(0)
numeric_cols['Followers'] = merged_df['Metadata'].str.extract(r'(\d+)\s+Follower').astype(float).fillna(0)
numeric_cols['Review_Length'] = merged_df['Review'].astype(str).apply(len)

plt.figure(figsize=(10, 8))
# Draw the heatmap
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Restaurant and Review Data')
plt.show()
````

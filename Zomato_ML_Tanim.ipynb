# **Project Summary** **-**

The modern restaurant industry operates in a highly competitive digital landscape where customer reviews and feedback play a pivotal role in shaping a business's reputation and financial success. Platforms like Zomato host massive volumes of unstructured textual data in the form of customer reviews. For restaurant owners and stakeholders, manually reading and analyzing thousands of reviews to gauge overall customer satisfaction is an impossible task. Therefore, the primary objective of this Capstone Project was to design, develop, and deploy an automated Natural Language Processing (NLP) machine learning pipeline capable of performing Sentiment Analysis on Zomato restaurant reviews, accurately classifying them into positive or negative/neutral sentiments.
The project commenced with a comprehensive Exploratory Data Analysis (EDA) phase to understand the underlying statistical distributions and correlations within the dataset. During this phase, statistical hypothesis testing was rigorously applied using Two-Sample T-Tests and Pearson Correlation Coefficients to mathematically validate assumptions regarding the relationships between restaurant costs, aggregate ratings, review lengths, and user engagement metrics like photo uploads.
Following the EDA, the unstructured data underwent an extensive data preprocessing and feature engineering phase. Standard tabular data cleaning included dropping missing values and capping numerical outliers in the cost feature using the IQR Winsorization method. The textual data required a meticulous NLP cleaning pipeline: expanding English contractions, converting text to lowercase, removing punctuation, filtering out URLs and numerical digits, and stripping out non-informative stopwords. To standardize the vocabulary, WordNet Lemmatization was applied to reduce words to their core dictionary roots. The cleaned text was then mathematically transformed into a machine-readable format using Term Frequency-Inverse Document Frequency (TF-IDF) Vectorization, strategically limited to the top 3,000 features to control dimensionality and prevent model overfitting.
A significant challenge identified in the dataset was extreme class imbalance, as customers are statistically more likely to leave highly positive reviews. To prevent the machine learning models from developing a predictive bias toward the majority class, the Synthetic Minority Over-sampling Technique (SMOTE) was applied exclusively to the training data to synthesize plausible minority class examples and create a perfectly balanced training environment.
Finally, the project transitioned into the predictive modeling phase. Three distinct classification algorithms were implemented and evaluated: Logistic Regression, Random Forest Classifier, and Multinomial Naive Bayes. Each model's performance was rigorously measured using Accuracy, Precision, Recall, and the F1-Score. To extract the absolute maximum predictive capability from the algorithms, hyperparameter optimization techniques, specifically GridSearchCV and RandomizedSearchCV, were deployed. After exhaustive cross-validation, the tuned Logistic Regression model was selected as the final production algorithm. It delivered the most balanced and robust F1-Score on unseen testing data while offering superior interpretability, lightweight computational demands, and rapid inference times. Ultimately, this deployed model empowers restaurant management to transition from reactive customer service to proactive quality control. By automatically flagging negative sentiment in real-time, businesses can immediately address service failures, improve menu offerings, and recover dissatisfied customers before their negative experiences impact overall revenue. The final pipeline was successfully serialized and exported, standing fully ready for seamless integration into a live web application.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# --- 1. CLEANING METADATA ---
# Fix the 'Cost' column: remove commas and turn it into a number
df_meta['Cost'] = df_meta['Cost'].astype(str).str.replace(',', '')
df_meta['Cost'] = pd.to_numeric(df_meta['Cost'])

# Fill missing values in Metadata with 'Not Available'
df_meta['Collections'] = df_meta['Collections'].fillna('Not Available')
df_meta['Timings'] = df_meta['Timings'].fillna('Not Available')

# --- 2. CLEANING REVIEWS ---
# Drop duplicate rows in Reviews
df_reviews = df_reviews.drop_duplicates()

# Drop rows that have missing values in Reviews
df_reviews = df_reviews.dropna()

# --- 3. VERIFY THE CLEANUP ---
print("Metadata missing values now:\n", df_meta.isnull().sum())
print("\nReviews missing values now:\n", df_reviews.isnull().sum())
print("\nReviews rows remaining after cleanup:", df_reviews.shape[0])
````

### What all manipulations have you done and insights you found?

Manipulations Done:

Data Type Conversion: I removed the commas from the Cost column in the Metadata dataset and converted it into a numeric data type so it can be used for calculations and charts.

Handling Missing Values (Imputation): I filled the missing values in the Collections and Timings columns with the string "Not Available" to ensure no blank spaces remained.

Handling Duplicates and Nulls (Deletion): In the Reviews dataset, I dropped the duplicate rows. I also dropped the rows containing missing values (nulls) because they represented a tiny fraction of the massive 10,000-row dataset, meaning their removal will not negatively impact our analysis.

Insights Found:

The datasets are now completely clean with 0 missing values and 0 duplicates. The Reviews dataset retained 9,955 clean rows. We are now ready to begin creating visual charts for Exploratory Data Analysis (EDA).

# ***5. Hypothesis Testing***

````python
# QUICK SETUP: Run this FIRST to load the data into this new notebook!
import pandas as pd
import scipy.stats as stats

# 1. Load the data
df_meta = pd.read_csv('Zomato Restaurant names and Metadata.csv')
df_reviews = pd.read_csv('Zomato Restaurant reviews.csv')

# 2. Clean Cost and Rating
df_meta['Cost'] = df_meta['Cost'].astype(str).str.replace(',', '')
df_meta['Cost'] = pd.to_numeric(df_meta['Cost'])
df_reviews = df_reviews.drop_duplicates().dropna()
df_reviews['Rating'] = pd.to_numeric(df_reviews['Rating'], errors='coerce')
df_reviews = df_reviews.dropna(subset=['Rating'])

# 3. Merge them into the 'merged_df' variable!
merged_df = pd.merge(df_reviews, df_meta, left_on='Restaurant', right_on='Name', how='inner')
print("Data is successfully loaded and ready for Hypothesis Testing!")
````

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.




Based on my Exploratory Data Analysis, here are three hypothetical statements to test:

Hypothesis 1 (Cost vs. Rating): Restaurants with higher ratings (4.0 and above) have a significantly different average cost compared to restaurants with lower ratings.

Hypothesis 2 (Cost vs. Pictures): There is a statistically significant correlation between the cost of a restaurant and the number of pictures uploaded by reviewers.

Hypothesis 3 (Review Length vs. Rating): There is a significant difference in the average length of a review when comparing 5-star ratings to 1-star ratings.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0): There is no significant difference in the average cost for two people between high-rated restaurants (>= 4.0) and low-rated restaurants (< 4.0).

Alternate Hypothesis (H1): There is a significant difference in the average cost for two people between high-rated and low-rated restaurants.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
import scipy.stats as stats

# Group the data into high-rated and low-rated costs
high_rated_cost = merged_df[merged_df['Rating'] >= 4.0]['Cost']
low_rated_cost = merged_df[merged_df['Rating'] < 4.0]['Cost']

# Perform a Two-Sample T-Test
t_stat, p_value = stats.ttest_ind(high_rated_cost, low_rated_cost, nan_policy='omit')

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Reject the Null Hypothesis. There is a significant difference in cost.")
else:
    print("Conclusion: Fail to reject the Null Hypothesis. There is no significant difference in cost.")
````

Which statistical test have you done to obtain P-Value?

I used an Independent Two-Sample T-Test.

##### Why did you choose the specific statistical test?

A Two-Sample T-Test is the correct statistical method for comparing the means of two independent groups (high-rated vs. low-rated restaurants) to determine if they are significantly different from each other.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0): There is no correlation between the cost of a restaurant and the number of pictures a user uploads in their review.

Alternate Hypothesis (H1): There is a significant correlation between the cost of a restaurant and the number of pictures uploaded.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# Drop any missing values in these two columns to ensure the test runs smoothly
clean_df = merged_df.dropna(subset=['Cost', 'Pictures'])

# Perform Pearson Correlation Test
corr_coefficient, p_value = stats.pearsonr(clean_df['Cost'], clean_df['Pictures'])

print(f"Pearson Correlation Coefficient: {corr_coefficient:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Reject the Null Hypothesis. There is a significant correlation.")
else:
    print("Conclusion: Fail to reject the Null Hypothesis. There is no significant correlation.")
````

##### Which statistical test have you done to obtain P-Value?

I used the Pearson Correlation Coefficient Test.

##### Why did you choose the specific statistical test?

The Pearson Correlation test is the standard statistical test used to evaluate the strength and direction of the linear relationship between two continuous numerical variables (Cost and Number of Pictures).

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0): There is no significant difference in the average text length of a 5-star review compared to a 1-star review.

Alternate Hypothesis (H1): There is a significant difference in the average text length between 5-star and 1-star reviews.

````python
# Perform Statistical Test to obtain P-Value
# Ensure we have the review length calculated
merged_df['Review_Length'] = merged_df['Review'].astype(str).apply(len)

# Isolate the review lengths for 5-star and 1-star reviews
five_star_len = merged_df[merged_df['Rating'] == 5.0]['Review_Length']
one_star_len = merged_df[merged_df['Rating'] == 1.0]['Review_Length']

# Perform a Two-Sample T-Test
t_stat, p_value = stats.ttest_ind(five_star_len, one_star_len, nan_policy='omit')

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Reject the Null Hypothesis. There is a significant difference in review length.")
else:
    print("Conclusion: Fail to reject the Null Hypothesis. There is no significant difference in review length.")
````

##### Which statistical test have you done to obtain P-Value?

I used an Independent Two-Sample T-Test.

##### Why did you choose the specific statistical test?

Because I am comparing the continuous variable of 'Review Length' across two distinctly separate and independent categorical groups (5-star ratings vs. 1-star ratings), the Two-Sample T-Test is the most appropriate way to test if their means are mathematically different.

````python
# Handling Outliers & Outlier treatments
# Traditional outlier treatment doesn't apply to text strings.
# However, for our numerical feature 'Cost', we cap extreme values using the IQR method.
Q1 = merged_df['Cost'].quantile(0.25)
Q3 = merged_df['Cost'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

# Cap outliers at the upper bound (Winsorization)
merged_df['Cost'] = merged_df['Cost'].apply(lambda x: upper_bound if x > upper_bound else x)
print(f"Numerical outliers in 'Cost' capped at: {upper_bound}")
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Text data does not have numerical outliers. However, for numerical features, I used Winsorization (IQR Method) to cap extreme values at the upper bound. I chose capping over deletion to ensure we do not lose valuable review text data just because a meal was unusually expensive.

#### 1. Expand Contraction

````python
# Expand Contraction
import re

# A basic dictionary of common contractions
contractions_dict = {"ain't": "is not", "aren't": "are not","can't": "cannot", "could've": "could have", "couldn't": "could not", "didn't": "did not",  "doesn't": "does not", "don't": "do not", "hadn't": "had not", "hasn't": "has not", "haven't": "have not", "he'd": "he would","he'll": "he will", "he's": "he is", "I'd": "I would", "I'll": "I will", "I'm": "I am", "I've": "I have", "isn't": "is not", "it's": "it is", "let's": "let us", "must've": "must have", "mustn't": "must not", "she'd": "she would", "she'll": "she will", "she's": "she is", "should've": "should have", "shouldn't": "should not", "that's": "that is", "there's": "there is", "they'd": "they would", "they'll": "they will", "they're": "they are", "they've": "they have", "wasn't": "was not", "we'd": "we would", "we'll": "we will", "we're": "we are", "we've": "we have", "weren't": "were not", "what's": "what is", "where's": "where is", "who's": "who is", "won't": "will not", "would've": "would have", "wouldn't": "would not", "you'd": "you would", "you'll": "you will", "you're": "you are", "you've": "you have"}

contractions_re = re.compile('(%s)' % '|'.join(contractions_dict.keys()))

def expand_contractions(text, contractions_dict=contractions_dict):
    def replace(match):
        return contractions_dict[match.group(0)]
    return contractions_re.sub(replace, str(text))

merged_df['Review'] = merged_df['Review'].apply(expand_contractions)
print("Contractions successfully expanded!")
````

I used SMOTE (Synthetic Minority Over-sampling Technique) strictly on the training data. Instead of randomly duplicating existing negative reviews (which causes overfitting) or deleting valuable positive reviews (which destroys data), SMOTE mathematically synthesizes completely new, plausible minority class examples by interpolating between existing data points. This ensures the model learns to identify both sentiments equally.

````python
# ML Model - 1 Implementation with GridSearchCV
from sklearn.model_selection import GridSearchCV

# Define the hyperparameters to tune
param_grid = {
    'C': [0.1, 1, 10],  # Regularization strength
    'solver': ['liblinear', 'lbfgs']
}

# Initialize GridSearchCV
grid_search = GridSearchCV(LogisticRegression(random_state=42, max_iter=2000), param_grid, cv=5, scoring='f1')

# Fit the Algorithm
grid_search.fit(X_train_resampled, y_train_resampled)
best_log_reg = grid_search.best_estimator_

# Predict on the test model
y_pred_log_tuned = best_log_reg.predict(X_test)

print(f"Best Hyperparameters: {grid_search.best_params_}")
````

##### Which hyperparameter optimization technique have you used and why?

I utilized GridSearchCV with 5-fold cross-validation. I chose this technique because it exhaustively searches through a predefined dictionary of hyperparameters (in this case, varying the regularization parameter 'C' and the 'solver' algorithm) to mathematically guarantee the absolute optimal combination for maximizing the F1-Score on the training data.

##### Have you seen any improvement? Note down the improvement with updates Evaluation metric Score Chart.

Yes, there was an improvement. By applying GridSearchCV, the model optimized its regularization parameter (C), which reduced overfitting on the SMOTE-resampled training data. This resulted in a slight increase in both the F1-Score and the overall testing accuracy compared to the baseline Logistic Regression model.

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 2 Implementation with RandomizedSearchCV
from sklearn.model_selection import RandomizedSearchCV

# Define hyperparameters to tune
param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# Initialize RandomizedSearchCV (using Randomized to save system RAM/Time)
random_search = RandomizedSearchCV(
    RandomForestClassifier(random_state=42),
    param_distributions=param_dist,
    n_iter=5,
    cv=3,
    scoring='f1',
    random_state=42,
    n_jobs=-1
)

# Fit the Algorithm
random_search.fit(X_train_resampled, y_train_resampled)
best_rf = random_search.best_estimator_

# Predict on the test model
y_pred_rf_tuned = best_rf.predict(X_test)

print(f"Best Hyperparameters: {random_search.best_params_}")
````

##### Which hyperparameter optimization technique have you used and why?

I utilized RandomizedSearchCV with 3-fold cross-validation. While GridSearchCV is exhaustive, training multiple Random Forests on 3,000 TF-IDF features is extremely computationally expensive. RandomizedSearchCV randomly samples a fixed number of parameter settings from the specified distributions, offering a highly efficient tradeoff that finds near-optimal hyperparameters (like tree depth and number of estimators) in a fraction of the time, preventing memory crashes.

##### Have you seen any improvement? Note down the improvement with updates Evaluation metric Score Chart.

Yes, there was an improvement in model generalization. Tuning the hyperparameters (specifically controlling max_depth and min_samples_split) helped constrain the depth of the individual decision trees. This prevented the Random Forest from memorizing the noise in the SMOTE-resampled training data, thereby reducing variance, lowering overfitting, and leading to a more robust F1-Score on the unseen test set.

#### 3. Explain each evaluation metric's indication towards business and the business impact pf the ML model used.

Accuracy: Indicates the overall percentage of reviews correctly classified. Business Impact: Gives stakeholders a high-level, easy-to-understand view of model reliability.

Precision: Out of all reviews predicted as 'Positive', how many were actually positive? Business Impact: High precision ensures we do not accidentally categorize angry reviews as positive, preventing false confidence in the restaurant's operational performance.

Recall: Out of all actual 'Positive' reviews, how many did the model successfully find? Business Impact: High recall ensures the business captures as much positive sentiment as possible to use for marketing materials and loyalty reward programs.

F1-Score: The harmonic mean of Precision and Recall. Business Impact: It provides a single, balanced metric for the business to evaluate the model's true effectiveness, which is crucial since our original review dataset was imbalanced.
